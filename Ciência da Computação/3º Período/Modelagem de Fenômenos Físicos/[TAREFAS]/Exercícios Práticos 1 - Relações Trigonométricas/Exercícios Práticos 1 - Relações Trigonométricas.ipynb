1. Um fotógrafo está posicionado em frente a uma plataforma de 
lançamento de foguetes de forma que o eixo da lente da sua câmera está a 
50 m de distância e perfeitamente alinhada com a base do foguete. Em que 
ângulo ele deve colocar a câmera para tirar uma foto deste foguete quando 
ele  estiver  há  200  m  de  altura.    Para  conseguir  os  pontos  desta  atividade 
você deverá (a) encontrar o ângulo 𝑼, usando a linguagem de programação 
Python, sem usar nenhuma função trigonométrica disponível na 
linguagem,  ou  em  suas  bibliotecas;  (b)  criar  uma  função  em  Python  que 
ajude o fotógrafo a ajustar o ângulo da câmera para qualquer altitude e (c) 
plotar, usando a biblioteca Matplotlib um diagrama do triângulo resultante 
em cada altitude.

```python
#1.a

import math

ca = 50 # Define o cateto adjacente ao ângulo (distância entre fotógrafo e foguete)
co = 200 # Define o cateto oposto ao ângulo (distância do fotógrafo ao chão)

alpha = math.atan(co / ca) # Calcula o ângulo
print(f"O ângulo é igual a {alpha:.3f} radianos ou a {math.degrees(alpha):.3f}°")
```

```python
#1.b

import math

def calcAlpha(dist, height):
  # Função para calcular o ângulo do fotógrafo
  # dist -> distância para a base do foguete
  # height -> altura do foguete do chão
  return math.atan(co / ca) # Calcula o ângulo


# Exemplo aplicado ao item anterior (distância sendo igual a 50 metros e a altura sendo igual a 200 metros)
print(calcAlpha(dist=50, height=200))
```

```python
#2.c

# OBS.: O SLIDER NÃO FUNCIONA NO GOOGLE COLAB (APENAS EXECUTANDO PELO PYTHON DIRETAMENTE)!!!

import matplotlib.pyplot as pyplot
import matplotlib.widgets as widgets
import matplotlib.lines as lines
import numpy as np

def getMinHeight(length):
  # Função para calcular altura mínima de acordo com o comprimento do telhado
  RATIO = 1/4
  return RATIO * length


length = 3 # Define o valor inicial do comprimento
rangeLength = [2, 7] # Define o intervalo para o slider do comprimento do telhado
chartsize = (6.5, 6.5) # Define o tamanho do gráfico

# Configuração inicial do gráfico
fig, ax = pyplot.subplots(figsize=chartsize)
fig.canvas.manager.set_window_title('Gráfico') # Define o nome da janela
ax.set_title(f"Gráfico da altura mínima para o comprimento") # Define o título do gráfico
ax.set_aspect('equal', 'datalim') # Obriga os eixos a serem proporcionais

ax.axhline(y=0, linestyle='--', color='black') # Coloca o eixo X no gráfico
ax.axvline(x=0, linestyle='--', color='black') # Coloca o eixo Y no gráfico

x = [0, length, length, 0] # Define as coordenadas no eixo X dos pontos do telhado
y = [0, 0, getMinHeight(length), 0] # Define as coordenadas no eixo Y dos pontos do telhado

ax.plot(x, y, linestyle='--', color='blue') # Cria uma máscara da posição inicial do telhado
lengthLine = lines.Line2D(x, y, color='blue') # Cria o telhado
ax.add_artist(lengthLine) # Coloca o telhado no gráfico

def update_lengthSlider(val):
  x = [0, val, val, 0] # Atualiza as coordenadas no eixo X dos pontos do telhado
  y = [0, 0, getMinHeight(val), 0] # Atualiza as coordenadas no eixo Y dos pontos do telhado
  lengthLine.set_data(x, y)

# Configuração do slider do comprimento
lengthSlider = widgets.Slider(pyplot.axes([0.3, 0.025, 0.6, 0.03]), 'Variar comprimento', valmin=min(rangeLength), valmax=max(rangeLength), valinit=length)
lengthSlider.on_changed(update_lengthSlider)

# Manobra para alinhar o gráfico com o maior triângulo possível
line, = ax.plot(max(rangeLength), getMinHeight(max(rangeLength)))
line.remove()

pyplot.show()
```

3. Usando o Python e as funções trigonométricas disponíveis nesta linguagem de programação 
e  suas  bibliotecas  crie  uma  função  que  receba  um  ângulo,  em  graus,  e  plote  um  circulo 
trigonométrico de raio 1, centrado na origem destacando este ângulo, seu seno, seu cosseno 
e sua tangente. 

```python
#3

import matplotlib.pyplot as pyplot
import numpy as np


def plotUnitCircle(angle):
  # Mostra o gráfico do círculo unitário com o ângulo dado (em graus) e indica
  # o seno, o cosseno e a tangente dele
  chartsize = (10, 10) # Define o tamanho do gráfico

  # Configuração inicial do gráfico
  fig, ax = pyplot.subplots(figsize=chartsize)
  fig.canvas.manager.set_window_title('Gráfico') # Define o nome da janela
  ax.set_title(f"Unit Circle with the angle {angle}°") # Define o título do gráfico
  ax.set_aspect('equal', 'datalim') # Obriga os eixos a serem proporcionais

  ax.axhline(y=0, linestyle='--', color='black') # Coloca o eixo X no gráfico
  ax.axvline(x=0, linestyle='--', color='black') # Coloca o eixo Y no gráfico
  ax.axvline(x=1, color='black') # Coloca o eixo da tangente no gráfico

  angle_rad = angle * np.pi / 180 # Converte o ângulo de graus para radianos

  alpha = np.linspace(0, 2 * np.pi, 500)
  ax.plot(np.cos(alpha), np.sin(alpha), color='blue') # Coloca um círculo unitário na origem no gráfico

  alpha = np.linspace(0, angle_rad, 500)
  ax.plot(0.1 * np.cos(alpha), 0.1 * np.sin(alpha), color='purple') # Coloca o semi-círculo do ângulo no gráfico

  sin = np.sin(angle_rad)
  cos = np.cos(angle_rad)
  tan = np.tan(angle_rad)
  ax.plot(np.linspace(cos, cos), np.linspace(0, sin), color='green', label='Seno') # Coloca a linha do seno no gráfico
  ax.plot(np.linspace(0, cos), np.linspace(0, 0), color='red', label='Cosseno') # Coloca a linha do cosseno no gráfico
  ax.plot(np.linspace(0, 1), np.linspace(0, tan), color='orange', label='Tangente') # Coloca a linha da tangente no gráfico
  ax.plot(np.linspace(0, cos), np.linspace(0, sin), linestyle='--', color='blue', label='Raio') # Coloca a linha do raio no gráfico
  ax.plot(cos, sin, marker='o', color='blue') # Coloca o ponto na circunferência pelo ângulo no gráfico
  ax.plot(1, tan, marker='o', color='orange') # Coloca o ponto na tangente pelo ângulo no gráfico
  ax.plot(0, 0, marker='o', color='black') # Coloca o ponto da origem no gráfico

  # Anotações no gráfico com indicações
  fig.gca().annotate('O', xy=(-0.075, -0.075), fontsize=10)
  fig.gca().annotate('Raio', xy=(cos / 2 - 0.05, sin / 2 + 0.05), fontsize=10, rotation=angle)
  fig.gca().annotate('Seno', xy=(cos + 0.025, sin / 3 + 0.05), fontsize=10, rotation=90)
  fig.gca().annotate('Cosseno', xy=(cos / 3,-0.075), fontsize=10)
  fig.gca().annotate('Tangente', xy=(2 / 3 - 0.05, 2 * tan / 3 + 0.05), fontsize=10, rotation=angle)

  ax.legend()
  pyplot.show()


# Exemplo para um ângulo de 60°
plotUnitCircle(angle=60)
```

4. Usando o Python e as funções trigonométricas disponíveis nesta linguagem de programação 
e suas bibliotecas crie uma função que receba um número inteiro 𝑛 tal que 1≤𝑛≤5 e plote 
as  funções  seno,  cosseno  e  tangente  para  todos  os  ângulos  possíveis  considerando  que 𝑛 
representa o número períodos completos.

```python
#4

import matplotlib.pyplot as pyplot
import matplotlib.widgets as widgets
import matplotlib.lines as lines
import numpy as np


def plotGraph(p):
  rangeP = [1, 5] # Define o intervalo para os períodos completos

  # Garantir que p está no intervalo adequado
  if p < min(rangeP):
    p = min(rangeP)
  elif p > max(rangeP):
    p = max(rangeP)

  chartsize = (6.5, 6.5) # Define o tamanho do gráfico

  # Configuração inicial do gráfico
  fig, ax = pyplot.subplots(figsize=chartsize)
  fig.canvas.manager.set_window_title('Gráfico') # Define o nome da janela
  ax.set_title(f"Gráfico das funções trigonométricas (com n={p})") # Define o título do gráfico
  ax.set_aspect('equal', 'datalim') # Obriga os eixos a serem proporcionais

  ax.axhline(y=0, linestyle='--', color='black') # Coloca o eixo X no gráfico
  ax.axvline(x=0, linestyle='--', color='black') # Coloca o eixo Y no gráfico

  x = np.linspace((p - 1) * np.pi * 2, p * np.pi * 2) # Define as coordenadas no eixo X
  sin = np.sin(x) # Define as coordenadas no eixo Y para o seno
  cos = np.cos(x) # Define as coordenadas no eixo Y para o cosseno
  tan = np.tan(x) # Define as coordenadas no eixo Y para o tangente

  ax.plot(x, sin, color='green', scaley=False) # Cria a função seno
  ax.plot(x, cos, color='orange', scaley=False) # Cria a função cosseno
  ax.plot(x, tan, color='red', scaley=False) # Cria a função tangente
  # OBS.: scaley=False foi usado para melhorar a visibilidade

  pyplot.show()


# Exemplo com 2 períodos completos
plotGraph(2)
```
